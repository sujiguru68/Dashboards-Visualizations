### GDP per Unit of Energy Use dataset

```python
gdp = pd.read_csv("C:/Users/sujik/OneDrive/Documents/MS - Data Science/Semester 3 (Fall_23)/DSC 530 - Data Visualization/Project/Final Presentation/DV_Final_Project_Files/gdp-per-unit-of-energy-use-2017-u.s.-dollar-ppp-per-kilogram-of-oil-equivalent-1.csv")
```

```python
gdp.info()
```

```python
print(gdp.isnull().sum())
```

```python
null_values = gdp[gdp.isnull().any(axis=1)]
print(null_values)
```

```python
gdp[gdp['Country']=='Bosnia and Herz.']
```

```python
gdp[gdp['Country']=='Belize']
```

```python
# Fill the first value with zero if it's null
gdp['1990'] = gdp['1990'].fillna(0)
```

```python
# Specify the range of columns for interpolation (from second to last)
columns_to_interpolate = gdp.columns[1:]

# Interpolate values in the specified columns for null values
gdp[columns_to_interpolate] = gdp[columns_to_interpolate].mask(gdp[columns_to_interpolate].isnull(), gdp[columns_to_interpolate].interpolate(axis=1))
```

```python
gdp[gdp['Country']=='Bosnia and Herz.']
```

```python
gdp[gdp['Country']=='Belize']
```

```python
gdp.to_csv('gdp-per-unit-of-energy-use-2017-u.s.-dollar-ppp-per-kilogram-of-oil-equivalent.csv', index=False)
```
